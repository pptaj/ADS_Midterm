```python
import pandas as pd
import numpy as np

#has NAs and Spaces. First replace NA by "unknown" and later spaces by "greater>65%" as per definition
df['ORIGINAL DEBT-TO-INCOME (DTI) RATIO'].fillna('Unknown', inplace=True)


#replacing empty by NaN - Not working 
df1 = df.replace(r'\s+', np.nan, regex=True)
#counting number of Nans in column
(df1['CREDIT SCORE'].isnull().sum())
#replace columnn1 "Credit Score" by Mean
#df1['CREDIT SCORE'].fillna(df1['CREDIT SCORE'].mean(), inplace=True)
#replace columnn1 "Credit Score" by Median
#df1['CREDIT SCORE'].fillna(df1['CREDIT SCORE'].median(), inplace=True)

#replace Nan in "Number of Borrowers" with 0 (19 values replaced with 0)
df1['NUMBER OF BORROWERS'].fillna('0', inplace=True)

#replace Nan in "Super Conforming Flag" with 'N'()(49920 values replaced with 'N')
df1['Super Conforming Flag'].fillna('N', inplace=True)

#replace all spaces in "Occupancy Status" with 'U'
#There are  no blanks?
df1['OCCUPANCY STATUS'] = df1['OCCUPANCY STATUS'].replace({' ':'U'})

#replace all spaces in "Channel" with 'Unknown'
df1['CHANNEL'].fillna("Unknown", inplace=True)

#split "first payment date" into two column-> "year and "month"
df1['FIRST PAYMENT YEAR']= df1['FIRST PAYMENT DATE'] // 100
df1['FIRST PAYMENT MONTH'] = df1['FIRST PAYMENT DATE'].astype(str).str[-2:].astype(np.int64)

#9th March
# replace NaN (7730) by Not Applicable 
df1['FIRST TIME HOMEBUYER FLAG'].fillna('Not Applicable', inplace=True)

#split Maturity date
df1['MATURITY YEAR']= df1['MATURITY DATE'] // 100
df1['MATURITY MONTH'] = df1['MATURITY DATE'].astype(str).str[-2:].astype(np.int64)

#Replace unknown METROPOLITAN STATISTICAL AREA (MSA) OR METROPOLITAN DIVISION  - 7729  
df1['METROPOLITAN STATISTICAL AREA (MSA) OR METROPOLITAN DIVISION'].fillna('Non-MSA/Unknown', inplace=True)

df1['MORTGAGE INSURANCE PERCENTAGE (MI %)'].fillna('Unknown', inplace=True)
#MORTGAGE INSURANCE PERCENTAGE doesnt seem to have  empty spaces. Please check?

#NUMBER OF UNITS doesnt seem to have  empty spaces. Please check?
df1['NUMBER OF UNITS'].fillna('Unknown', inplace=True)

#Ocuupancy Status
df1['OCCUPANCY STATUS'].fillna('Unknown', inplace=True)

#"ORIGINAL COMBINED LOAN-TO-VALUE (CLTV)"
df1["ORIGINAL COMBINED LOAN-TO-VALUE (CLTV)"].fillna('Unknown', inplace=True)

#ORIGINAL DEBT-TO-INCOME (DTI) RATIO - 56 unknowns, 1031 values replaced by "Greater than 65%
df1['ORIGINAL DEBT-TO-INCOME (DTI) RATIO'].fillna("Greater than 65%", inplace=True)

#ORIGINAL UPB - Replace by NA - No NaN here
df1['ORIGINAL UPB'].fillna("Unknown" ,inplace=True)

#ORIGINAL LOAN-TO-VALUE (LTV)
df1['ORIGINAL LOAN-TO-VALUE (LTV)'].fillna("Unknown" ,inplace=True)

#ORIGINAL INTEREST RATE
df1['ORIGINAL INTEREST RATE'].fillna("Unknown", inplace=True)

#PREPAYMENT PENALTY MORTGAGE (PPM) FLAG - 1039 values changed
df1['PREPAYMENT PENALTY MORTGAGE (PPM) FLAG'].fillna("Unknown" ,inplace=True)

#PROPERTY STATE
df1['PROPERTY STATE'].fillna("Unknown" ,inplace=True)

#POSTAL CODE
df1['POSTAL CODE'].fillna("Unknown" ,inplace=True)

#LOAN PURPOSE
df1['LOAN PURPOSE'].fillna("Unknown" ,inplace=True)

#ORIGINAL LOAN TERM - Data dictionary has conditions..Work on that??
df1['ORIGINAL LOAN TERM'].fillna("Unknown" ,inplace=True)

#NUMBER OF BORROWERS - 19 unknown
df1['NUMBER OF BORROWERS'].fillna("Unknown" ,inplace=True)

#SELLER NAME
df1['SELLER NAME'].fillna("Other Sellers" ,inplace=True)

#SERVICER NAME
df1['SERVICER NAME'].fillna("Other Servicers" ,inplace=True)

#Super Conforming Flag - 49920
df1['Super Conforming Flag'].fillna("N" ,inplace=True)


#print(df1)

```

```python

#SUMMARY METRICS

# Grouping by Seller Name and Number of Borrowers to get a list of unique Seller Name and Number of Borrowers pair
d= pd.DataFrame({'count' : df1.groupby( [ "SELLER NAME", "NUMBER OF BORROWERS"] ).size()}).reset_index()
print(d)
# Grouping by Seller Name to get the count of unique sellers
d1= pd.DataFrame({'unique_number_of_sellers' : d.groupby( ["SELLER NAME"] ).size()}).reset_index()
print(d1)     
        
#Count of borrowers for seller        
d2 = df1.groupby(['SELLER NAME'])['NUMBER OF BORROWERS'].count().reset_index(name="Number of Borrowers per seller")
print(d2) 

#Count of Occupancy Status        
d3 = df1.groupby(['OCCUPANCY STATUS'])['OCCUPANCY STATUS'].count().reset_index(name="Number of Occupancy Status")
print(d3) 

#Count of Super Conforming Flag       
d4 = df1.groupby(['Super Conforming Flag'])['Super Conforming Flag'].count().reset_index(name="Count of Super Conforming Flag ")
print(d4) 

#Top 10 Original UPB       
d5 = df1.groupby(['ORIGINAL UPB'])['ORIGINAL UPB'].count().head(10).reset_index(name="Count of Original UPB").sort_values(['Count of Original UPB'], ascending=False)
print(d5) 

#Top 10 most used Servicer Name       
d6 = df1.groupby(['SERVICER NAME'])['SERVICER NAME'].count().head(10).reset_index(name="Top 10 SERVICER NAME").sort_values(['Top 10 SERVICER NAME'], ascending=False)
print(d6)

#Top 10 most used Seller Name       
d7 = df1.groupby(['SELLER NAME'])['SELLER NAME'].count().head(10).reset_index(name="Top 10 SELLER NAME").sort_values(['Top 10 SELLER NAME'], ascending=False)
print(d7)

# Retrieving Unique Number of property State
d9= pd.DataFrame({'unique_number_of_property_state' : df1.groupby( ["PROPERTY STATE"] ).size()}).reset_index()
print(d9)  

# Retrieving Top 10 property States based on popularity
d10 = df1.groupby(['PROPERTY STATE'])['PROPERTY STATE'].count().head(10).reset_index(name="Top 10 Property State").sort_values(['Top 10 Property State'], ascending=False)
print(d10)  

# Retrieving First time Home Buyers
d11 = df1.groupby(['FIRST TIME HOMEBUYER FLAG'])['FIRST TIME HOMEBUYER FLAG'].count().reset_index(name="First time Home Buyers").sort_values(['First time Home Buyers'], ascending=False)
print(d11)  
```
